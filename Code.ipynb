#### ii) Centralize the network
    Remove the friends of friends from the network

```python
def centralize(friend_network:nx.Graph, my_id:str = 'bilalansari.000') -> nx.Graph:
    my_friends = list(friend_network.neighbors(my_id))
    for friend in my_friends:
        matualfriends = np.intersect1d(my_friends, list(friend_network.neighbors(friend)))
        non_matualfriends = np.setdiff1d(list(friend_network.neighbors(friend)), matualfriends)
        friend_network.remove_nodes_from(non_matualfriends)
    friend_network.remove_node('bilalansari.000')
    return friend_network.to_undirected()
```

```python
friend_network = centralize(friend_network)
```

#### iii) Remove the friends having matual friends less than 4

```python
def removeNodes(friend_network:nx.Graph, deg: int = 4) -> nx.Graph:
    remove = [node for node,degree in dict(friend_network.degree()).items() if degree < deg]
    friend_network.remove_nodes_from(remove)
    return friend_network
```

```python
friend_network = removeNodes(friend_network)
```

```python
#Save the Preprocessed Network
nx.write_gml(friend_network, 'data/preprocessed_network.gml')
```

### 3. Network Summary
    Size: number of edges in the network
    Order: number of nodes in the network
    Diameter: maximum eccentricity
    Average Clustering coefficient: mean of local clusterings coefficient
    Clustering coefficient: For unweighted graphs, the clustering of a node u is the fraction of possible triangles through that node that exist.

```python
def get_attr_name(friend_network:nx.Graph):
    return set([attr for friend in friend_network.nodes for attr in friend_network.nodes[friend].keys()])
```

```python
def get_diameter(friend_network:nx.Graph):
    """Return Giant Component Diameter"""
    giant_comp = max(nx.connected_component_subgraphs(friend_network), key=len)
    return nx.diameter(giant_comp)
```

```python
net_summary =  [friend_network.size(), friend_network.order(), get_attr_name(friend_network),
            nx.number_connected_components(friend_network), 
                nx.average_clustering(friend_network), 
               get_diameter(friend_network), 
                nx.density(friend_network), np.min(list(dict(friend_network.degree()).values()))]
index = ['Size','Order','Node Attributes','No. of Connected Components', 'Average Clustering coefficient', 
         "Giant Component's Diameter", 'Density', 'Average Degree']
pd.DataFrame({"Network Summary":net_summary} , index=index)
```

#### v) Node Similarity

```python
import scipy.spatial as spt
friend_matrix = nx.adjacency_matrix(friend_network).todense()
friend_matrix = nx.to_numpy_matrix(friend_network)

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(10,10))
ax1.imshow(friend_matrix, cmap = 'Greys', interpolation = 'None')
ax1.set_title('Adjacency Matrix')

D = np.corrcoef(friend_matrix)
ax2.imshow(D, cmap = 'Greys', interpolation = 'None')
ax2.set_title('Correlation coeff.')

dVec = spt.distance.pdist(friend_matrix, metric = 'euclidean')
D = spt.distance.squareform(dVec)
ax3.imshow(D, cmap = 'Greys', interpolation = 'None')
ax3.set_title('Euclidean Dist.')

dVec = spt.distance.pdist(friend_matrix, metric = 'cosine')
D = spt.distance.squareform(dVec)
ax4.imshow(D, cmap = 'Greys', interpolation = 'None')
ax4.set_title('Cosine Dist.')
plt.show()
```

```python
#Descriptive statistics
def get_descriptive_stat(graph:nx.Graph)->dict:
    stat = {}
    degree_sequence = list(dict(graph.degree()).values())
    stat['Number of nodes'] = len(graph.nodes())
    stat['Number of edges'] = len(graph.edges())
    stat['Maximum degree'] = np.max(degree_sequence)
    stat['Minimum degree'] = np.min(degree_sequence)
    stat['Std degree'] = np.std(degree_sequence)
    stat['Average degree'] = np.mean(degree_sequence)
    stat['Median degree'] = np.median(degree_sequence)
    return stat
```

```python
data = {}
for name,graph in [('My SN',friend_network),('Erdos-Rényi',graph_erdos),('Barabasi-Albert',graph_barabasi)]:
    stat = get_descriptive_stat(graph)
    index = list(dict(stat).keys())
    data[name] = list(dict(stat).values())
pd.DataFrame(data, index = index, dtype=int).T
```

### 5. Community Detection

#### i) Clique Search

```python
#Find the Maximum Clique
from networkx.algorithms.approximation import clique
maximum_clique = nx.algorithms.approximation.clique.max_clique(friend_network)
```

```python
print("Size of maximum clique: ", len(maximum_clique))
```

```python
#Add attributes to graph based on the maximum clique
node_size = dict(map(lambda node: (node,50) if node in maximum_clique else (node,20),list(friend_network.nodes)))
edge_weight = dict(map(lambda e: ((e[0],e[1]),10) if e[0] in maximum_clique and e[1] in maximum_clique else ((e[0],e[1]),4),list(friend_network.edges)))
for node in friend_network.nodes:
    friend_network.nodes[node]['nodeSize'] = node_size[node]
for u,v in friend_network.edges:
    friend_network.edges[u,v]['weight'] = edge_weight[(u,v)]
```

```python
nx.write_gml(friend_network,'data/maximum_clique_graph.gml')
```
